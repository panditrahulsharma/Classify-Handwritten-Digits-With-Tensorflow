```
from keras.datasets import mnist
from keras.utils import to_categorical

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```
test_images[0].shape

```

```
type(test_images[0])
```

```
#convert data into the min max scalling
#here image pixel max=255 min=0 so we can direct write=x/255
train_images=train_images.astype('float32') / 255
test_images=test_images.astype('float32')/255
```

```
#keras is a library to convert label encode data into onhotencoder
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
#method in caras convert catagorical data into onhot encoder
#in labels data is 0...9 so convert it
```

```
test_labels.shape
#(60000, 10) type only
```

```
from keras import layers
from keras import models
#from keras.layers import Conv2D
```

```
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(16, (5,5), activation = "relu"))
model.add(layers.MaxPooling2D((2, 2)))
#(28,28,1) 1-means black and white data
#in first layer add 32 nodes by 3*3 matrix actiovation='relu' input_shape=(28,28,1)
#max pool layer 2*2 metrix
#again convolution16 and 5*5 metrics
#again max polling 2*2
```

```
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
#model.add 64 is hidden layer of neural network and second layer is output layer 10 is ouput
model.add(layers.Dense(10, activation='softmax'))
#here in dense input shape not given because faltten work as a input shape
#actiovation softmax because not a binary or single input when multiple output use softmax if binary then use sigmoid
```

```
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.fit(train_images, train_labels, epochs=50, batch_size=64)

test_loss, test_acc = model.evaluate(test_images, test_labels)
```

```
print("test_loss=",test_loss)
print("test_acc=",test_acc)
```

```
prediction=model.predict(test_images)
```
